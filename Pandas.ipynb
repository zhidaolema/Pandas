```python
#Pandas处理丢失数据

import pandas as pd 
import numpy as np
dates = pd.date_range('20250515',periods=6)
df = pd.DataFrame(np.arange(24).reshape((6,4)),index=dates,columns=['A','B','C','D'])
df.iloc[0,1]=np.nan
df.iloc[1,2]=np.nan
print(df)
print(df.dropna(axis=0,how='any'))#how = {any,（all：所扫描的行or列值值全为nan，这行or列才会丢弃但并未删除）} axis=0 竖直方向对行进行操作 =1横向方向对列进行操作
#df.fillna(value=0) 将nan 值填充为
print(df.isnull())#打印整体的数据缺失Ture/False情况 说明数据是否缺失
print(np.any(df.isnull())==True)#判断表中是否有缺失值
```

```python
#Pandas 合并
import numpy as np
import pandas as pd
df1 = pd.DataFrame(np.ones((3,4))*0,columns=['a','b','c','d'])
df2 = pd.DataFrame(np.ones((3,4))*1,columns=['a','b','c','d'])
df3 = pd.DataFrame(np.ones((3,4))*2,columns=['a','b','c','d'])


#Concatenating
res = pd.concat([df1,df2,df3],axis=0,ignore_index=True)#0竖向合并   1横向合并
#print(res)

#join,['inner','outer']
df4 = pd.DataFrame(np.ones((3,4))*0,columns=['a','b','c','d'],index=[1,2,3])
df5 = pd.DataFrame(np.ones((3,4))*1,columns=['b','c','d','e'],index=[2,3,4])
res2 = pd.concat([df4,df5],join='inner',ignore_index=True)#取两者交集
res3 = pd.concat([df4,df5],join='outer',ignore_index=True)#取两者并集 没有的数据填充Nan
print(res2)
print(res3)

#merging two df by key/keys.(may be used in database)
#1.simple example
left = pd.DataFrame({'key':['K0','K1','K2','K3'],
                       'A':['A0','A1','A2','A3'],
                       'B':['B0','B1','B2','B3']})
right = pd.DataFrame({'key':['K0','K1','K2','K3'],
                       'C':['C0','C1','C2','C3'],
                       'D':['D0','D1','D2','D3']})
print(right)
print(left)
res4 = pd.merge(left,right,on='key')
print(res4)


#2.consider two keys
left = pd.DataFrame({'key1':['K0','K0','K1','K2'],
                     'key2':['K0','K1','K0','K1'],
                       'A':['A0','A1','A2','A3'],
                       'B':['B0','B1','B2','B3']})
right = pd.DataFrame({'key1':['K0','K1','K1','K2'],
                      'key2':['K0','K0','K0','K0'],
                       'C':['C0','C1','C2','C3'],
                       'D':['D0','D1','D2','D3']})
print(left)
print(right)
res5 = pd.merge(left,right,on=['key1','key2'],how='inner')#默认inner交  how=['left','right','inner','outer']
print(res5)


```

```python
#3.indicator
import numpy as np
import pandas as pd

df1 = pd.DataFrame(np.ones((3,4))*0,columns=['a','b','c','d'])
df2 = pd.DataFrame(np.ones((3,4))*1,columns=['a','b','c','d'])
df3 = pd.DataFrame(np.ones((3,4))*2,columns=['a','b','c','d'])


#Concatenating
res = pd.concat([df1,df2,df3],axis=0,ignore_index=True)#0竖向合并   1横向合并
#print(res)

#join,['inner','outer']
df4 = pd.DataFrame(np.ones((3,4))*0,columns=['a','b','c','d'],index=[1,2,3])
df5 = pd.DataFrame(np.ones((3,4))*1,columns=['b','c','d','e'],index=[2,3,4])
res2 = pd.concat([df4,df5],join='inner',ignore_index=True)#取两者交集
res3 = pd.concat([df4,df5],join='outer',ignore_index=True)#取两者并集 没有的数据填充Nan
print(res2)
print(res3)

```

```python
#merging two df by key/keys.(may be used in database)
#simple example
import numpy as np
import pandas as pd

left = pd.DataFrame({'key':['K0','K1','K2','K3'],
                       'A':['A0','A1','A2','A3'],
                       'B':['B0','B1','B2','B3']})
right = pd.DataFrame({'key':['K0','K1','K2','K3'],
                       'C':['C0','C1','C2','C3'],
                       'D':['D0','D1','D2','D3']})
#print(right)
#print(left)
res4 = pd.merge(left,right,on='key')
#print(res4)

df6 = pd.DataFrame({'coll':[0,1],'col_left':['a','b']})
df7 = pd.DataFrame({'coll':[1,2,2],'col_right':[2,2,2]})
print(df6)
print(df7)
#res5 = pd.merge(df6,df7,on='coll',how='outer',indicator=True) #outer 合并后没有元素进行填充 indicator默认False, 显示根据哪个列名合并
#print(res5)

#give the indicator a custom name
res6 = pd.merge(df6,df7,on='coll',how='outer',indicator='indicator_column')
print(res6)

#merge by index
left = pd.DataFrame({'A':['A0','A1','A2'],
                     'B':['B0','B1','B2']},
                     index=['K0','K1','K2'])
right = pd.DataFrame({'C':['C0','C1','C2'],
                     'D':['D0','D1','D2']},
                     index=['K0','K2','K3'])
print(left)
print(right)
#left_index and right_index
res7 = pd.merge(left,right,left_index=True,right_index=True,how='outer')
print(res7)
res8 = pd.merge(left,right,left_index=True,right_index=True,how='inner')
print(res8)

#handle overlapping
boys = pd.DataFrame({'k':['K0','K1','K2'],'age':[1,2,3]})#两组数据中age名字相同内涵不同，做区分 利用suffixes=后缀 改名为age_boys age_girls
girls = pd.DataFrame({'k':['K0','K0','K3'],'age':[4,5,6]})
res9 = pd.merge(boys,girls,on='k',suffixes=['_boys','_girls'],how='inner')
print(res9)
```
